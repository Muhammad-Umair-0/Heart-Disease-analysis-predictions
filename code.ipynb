```python
column = ['slope','thalach','restecg', 'cp', 'target']
df = df[column]
```

```python
plt.hist(df['thalach'], bins = 20, rwidth=0.8, density = True)
plt.xlabel('thalach')
plt.ylabel('count')
plt.show()
```

```python
df.shape
```

```python
Q1 = df.thalach.quantile(0.25)
Q3 = df.thalach.quantile(0.75)
IQR = Q3-Q1
```

```python
upper_limit = Q3+1.5*IQR
lower_limit = Q1-1.5*IQR
upper_limit,lower_limit
```

```python
df[(df.thalach >upper_limit) | (df.thalach <lower_limit)]
```

```python
df1 = df[(df.thalach < upper_limit) & (df.thalach > lower_limit)]
```

```python
plt.hist(df1['thalach'], bins = 20, rwidth=0.8, density = True)
plt.xlabel('thalach')
plt.ylabel('count')
plt.show()
```

```python
print(df1.cp.unique())
print(df1.restecg.unique())
print(df1.slope.unique())
```

```python
multi_cat = ['slope','restecg', 'cp']
```

```python
df1 = pd.get_dummies(df1, dtype=int ,columns=multi_cat,drop_first=True)
df1
```

```python
cont_val = ['thalach']
```

```python
from sklearn.preprocessing import StandardScaler
st = StandardScaler()
df1[cont_val] = st.fit_transform(df1[cont_val])
```

```python
df1
```

```python
from sklearn.model_selection import train_test_split
X = df1.drop('target', axis=1)
y = df1['target']
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=40)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
# importing model
from sklearn.linear_model import LogisticRegression

Log_reg = LogisticRegression()
Log_reg.fit(X_train, y_train)
```

```python
y_pred = Log_reg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
